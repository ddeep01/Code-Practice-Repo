## Autograd 

Step 1 : Make requires_grad=True (default=False) --> PyTorch starts tracking everything done with tensor.

Step 2 : backward() --> computes gradients of the final scalar output.

Step 3 : x.grad  --> stores gradient value and  Only leaf tensors have .grad .

```python
x = torch.tensor(2.0, requires_grad=True)
w = torch.tensor(3.0, requires_grad=True)

y = w * x
y.backward()

print(x.grad)
print(w.grad)
```

## Create Simple Neural Network

```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.shape
```

```python
df.drop(["id", "Unnamed: 32"], axis=1, inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
y_train
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
###  Numpy arrays to PyTorch tensors

X_train_tensor = torch.from_numpy(X_train)
X_test_tensor = torch.from_numpy(X_test)
y_train_tensor = torch.from_numpy(y_train)
y_test_tensor = torch.from_numpy(y_test)
```

```python
print(X_train_tensor.shape)
```

```python
class SimpleNN():

  def __init__(self, X):

    self.weights = torch.rand(X.shape[1], 1, dtype=torch.float64, requires_grad=True)
    self.bias = torch.zeros(1, dtype=torch.float64, requires_grad=True)

  def forward(self, X):
    z = torch.matmul(X, self.weights) + self.bias
    y_pred = torch.sigmoid(z)
    return y_pred

  def loss_function(self, y_pred, y):
    # Clamp predictions to avoid log(0)
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)

    # Calculate loss
    loss = -(y_train_tensor * torch.log(y_pred) + (1 - y_train_tensor) * torch.log(1 - y_pred)).mean()
    return loss
```

```python
learning_rate = 0.1
epochs = 25
```

```python
model = SimpleNN(X_train_tensor)

for epoch in range(epochs):

    # Forward Pass
    y_pred = model.forward(X_train_tensor)

    # loss calculate
    loss = model.loss_function(y_pred, y_train_tensor)

    # Backpropagation
    loss.backward()

    # Update Paramter
    with torch.no_grad():
        model.weights -= learning_rate * model.weights.grad
        model.bias -= learning_rate * model.bias.grad

    
    # zero gradients
    model.weights.grad.zero_()
    model.bias.grad.zero_()

    # print loss in each epoch
    print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
```

```python
model.bias
```

```python
# model evaluation
with torch.no_grad():
  y_pred = model.forward(X_test_tensor)
  y_pred = (y_pred > 0.9).float()
  accuracy = (y_pred == y_test_tensor).float().mean()
  print(f'Accuracy: {accuracy.item()}')
```

## torch.nn  (PyTorchâ€™s neural network toolbox.)

It provides:

Layers (Linear, Conv, LSTM, etc.)

Activation functions (ReLU, Sigmoidâ€¦)

Loss functions (MSE, CrossEntropyâ€¦)

The base class nn.Module (heart of PyTorch models)

ðŸ‘‰ Every neural network in PyTorch is a class that inherits from nn.Module.

```python
import torch
import torch.nn as nn
```

```python
class SimpleNNModule(nn.Module):

    def __init__(self,features):
        super().__init__()
        self.liner=nn.Linear(features,1)
        self.sigmoid=nn.Sigmoid()
    
    def forward(self,features):
        return self.sigmoid(self.liner(features))
```

```python
X=torch.rand(10,5)

model=SimpleNNModule(X.shape[1])
print(model)
model(X)
```

```python
model.liner.weight
```

```python
model.liner.bias
```

```python
class SimpleNNModule(nn.Module):

    def __init__(self,features):
        super().__init__()
        self.layer1=nn.Linear(features,3)
        self.relu=nn.ReLU()
        self.layer2=nn.Linear(3,1)
        self.sigmoid=nn.Sigmoid()
    
    def forward(self,features):

        out=self.layer1(features)
        out=self.relu(out)
        out=self.layer2(out)
        out=self.sigmoid(out)

        return out
```

```python
features=torch.rand(10,5)
model = SimpleNNModule(features.shape[1])
y_pred=model(features)
print(y_pred)
print()
print(model.layer1.weight)
print()
print(model.layer2.weight)
```

```python
class MySimpleNNModule(nn.Module):

    def __init__(self,features):
        super().__init__()
        self.network=nn.Sequential(nn.Linear(features,3),
                                   nn.ReLU(),
                                   nn.Linear(3,1),
                                   nn.Sigmoid()
                                   )
    
    def forward(self,features):
        return self.network(features)
```

```python
features=torch.rand(10,5)
model = MySimpleNNModule(features.shape[1])
y_pred=model(features)
print(y_pred)
print()
print("First layer weights:\n", model.network[0].weight)
print("First layer bias:\n", model.network[0].bias)

print("\nSecond layer weights:\n", model.network[2].weight)
print("Second layer bias:\n", model.network[2].bias)
```

## torch.nn on Real World Dataset

```python
import numpy as np
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
df.drop(["id","Unnamed: 32"],axis=1, inplace=True)
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,1:],df.iloc[:, 0], test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
X_train_tensor=torch.from_numpy(X_train)
X_test_tensor=torch.from_numpy(X_test)
y_train_tensor=torch.from_numpy(y_train)
y_test_tensor=torch.from_numpy(y_test)

# Ensure that Always datatype of all tensor are same 
# nn.Linear(...)   â†’ parameters are float32 by default
# torch.from_numpy(X_train) â†’ float64

X_train_tensor=X_train_tensor.to(torch.float32)
X_test_tensor=X_test_tensor.to(torch.float32)
y_train_tensor=y_train_tensor.to(torch.float32)
y_test_tensor=y_test_tensor.to(torch.float32)

y_train_tensor = y_train_tensor.view(-1, 1)   #  ypred shape is [455,1] but y_train_tensor shape is [455,] so we need to convert them in same shape  (-1,1)-->Columns Vector and -1 means: â€œPyTorch, you figure out this dimension for me.â€
y_test_tensor = y_test_tensor.view(-1, 1)
```

```python
class NeuralNetwork(nn.Module):

    def __init__(self,features):

        super().__init__()
        self.neural_network=nn.Sequential(
            nn.Linear(features,5),
            nn.ReLU(),
            nn.Linear(5,1),
            # nn.Sigmoid()            # We are not using Sigmoid here because BCEWithLogitsLoss already applies Sigmoid internally.
        )

    def forward(self,features):
        return self.neural_network(features)
```

```python
model = NeuralNetwork(X_train_tensor.shape[1])

optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
loss_fn = nn.BCEWithLogitsLoss()

for epoch in range(100):

    optimizer.zero_grad()      #  Reset Gradients
    y_pred = model(X_train_tensor)   # Forward Pass
    loss = loss_fn(y_pred, y_train_tensor)  # Loss
    loss.backward()           # Backpropagation
    optimizer.step()           # Update

    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")

```

```python
def binary_accuracy(logits, y_true):
    probs = torch.sigmoid(logits)
    preds = (probs >= 0.5).float()
    acc = (preds == y_true).float().mean()
    return acc
```

```python
model.eval()
with torch.no_grad():
    train_logits = model(X_train_tensor)
    train_acc = binary_accuracy(train_logits, y_train_tensor)

print(f"Train Accuracy: {train_acc.item()*100:.2f}%")
```

```python
model.eval()
with torch.no_grad():
    test_logits = model(X_test_tensor)
    test_acc = binary_accuracy(test_logits, y_test_tensor)

print(f"Test Accuracy: {test_acc.item()*100:.2f}%")
```
